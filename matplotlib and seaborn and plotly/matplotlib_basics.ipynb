```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 11)
y = 2 * x + 5
```

```python
# plot x and y
plt.plot(x, y)
plt.xlabel('numbers 1 to 10')
plt.ylabel('y = 2x + 5')
plt.title('Simple Plot')
```

```python
# xlimit and ylimit
plt.plot(x, y)
plt.xlim(0, 15)
plt.ylim(0, 40)
```

```python
#x = np.linspace(0, 100, 50)
x = np.arange(0, 100, 10)
y = x ** 4
x
```

```python
figure = plt.figure(figsize=(12, 8),dpi=300)
plt.scatter(x, y)
plt.plot(x, y, 'r--')  # red dashed line
plt.xticks(np.arange(0, 110, 5))
```

```python
fig2 = plt.figure()
axis1 = fig2.add_axes([0.1, 0.1, 1, 1])
axis2 = fig2.add_axes([0.2, 0.5, 0.4, 0.3])
axis1.plot(x, y)
axis1.set_title('Larger Plot')
axis2.plot(x[5:10], y[5:10], 'r--')
```

```python
x = np.linspace(0, 10, 11)
y = x ** 2
```

```python
figure = plt.figure(figsize=(10, 5),dpi=200)
plt.subplot(1, 2, 1)
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.subplot(1, 2, 2)
plt.plot(x, y, 'r--')
plt.title('Line Plot')
```

```python
x1 = np.linspace(0, 10, 11) # start 0 end 10 num of points 11
x2 = np.linspace(0, 20, 11) # start 0 end 20 num of points 11
x3 = np.linspace(0, 30, 11)
x4 = np.linspace(0, 40, 11)
y1 = x1 ** 2
y2 = x2 ** 4
y3 = np.sin(x3)
y4 = x4 ** 0.5
```

```python
figure = plt.figure(figsize=(10, 8),dpi=200)

plt.subplot(2, 2, 1) # row 1, col 1, plot 1
plt.plot(x1, y1, 'r--')
plt.title('y = x^2')

plt.subplot(2, 2, 2) # row 1, col 2, plot 2
plt.plot(x2, y2, 'g-')
plt.title('y = x^4')

plt.subplot(2, 2, 3) # row 2, col 1, plot 3
plt.plot(x3, y3, 'bx')
plt.title('y = sin(x)')

plt.subplot(2, 2, 4) # row 2, col 2, plot 4
plt.plot(x4, y4, 'y.')
plt.title('y = sqrt(x)')


```

```python
x = np.arange(1, 11, 0.1) # start stop and step size
```

```python
ysine = np.sin(x)
yclose = np.cos(x)
```

```python
plt.plot(x, ysine, 'r--', label='sine')
plt.plot(x, yclose, 'g-', label='cosine')
# move legend to upper left
plt.legend(loc='upper left')
```

```python
plt.scatter(x, ysine,c='blue', marker='x', s=10, alpha=0.1)
plt.scatter(x, yclose, marker='+', s=10, color = "#FF0000")
plt.legend(['sine', 'cosine'])
```

```python
x = np.arange(1,11)
y = x + 2
z = x + 4
```

```python
plt.plot(x,color='red', linestyle='--', marker='o', markersize=5)
plt.plot(y,color='blue', linestyle='-.', marker='x', markersize=8)
plt.plot(z,color='green', linestyle=':', marker='^', markersize=10)
```

```python
# line width
plt.plot(x, color ='purple', linewidth=5)
plt.plot(y, 'r--', linewidth=1)
plt.plot(z, 'g:', linewidth=10)
```

```python
# dotted line 
fig,ax = plt.subplots()
line = ax.plot(x, y, 'r:', linewidth=2)
line[0].set_dashes([5, 2, 10, 5]) # 5 points on, 2 off, 10 on, 5 off
```
